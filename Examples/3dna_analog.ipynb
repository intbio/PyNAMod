### Instead of using in-build pair finder function user may provide pairs list

```python
# Sample pairs list generation
u = mda.Universe(io.StringIO(pypdb.get_pdb_file('1KX5')), format='PDB')
residsI = u.select_atoms('segid I and nucleic').residues.resids[::10]
residsJ = u.select_atoms('segid J and nucleic').residues.resids[::-10]
pairs_list = []
for resid1,resid2 in zip(residsI,residsJ):
    pairs_list.append(('I',resid1,'J',resid2))
```

```python
pairs_with_pairs_list,params_with_pairs_list = analyze_structure(pdb_id='1KX5',leading_strands=['I'],pairs_list=pairs_list)
```
